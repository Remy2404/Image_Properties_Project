```python
import sys
import os
sys.path.append(os.path.abspath('..'))
from src.image_utils import load_image, show_image_matplotlib, get_image_properties, save_image

# Load images
image_paths = ["../data/sigma_cat.png", "../data/uiia_cat.png"]
output_paths = ["../results/processed_sigma_cat_phon_Ramy.png", "../results/uiia_results_cat_phon_Ramy.png"]

for image_path, output_path in zip(image_paths, output_paths):
	if not os.path.exists(image_path):
		print(f"Image not found at {image_path}")
		continue
	img = load_image(image_path)

	# Display image using Matplotlib
	show_image_matplotlib(img)

	# Get image properties
	shape, dtype, pixel_value = get_image_properties(img)

	print(f"Image Path: {image_path}")
	print(f"Image Shape: {shape}")
	print(f"Image Data Type: {dtype}")
	print(f"Pixel Value at (0,0): {pixel_value}")

	# Save processed image
	save_image(img, output_path)
```
